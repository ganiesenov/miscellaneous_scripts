## Задание 1: Модель преобразованной зависимой переменной

Дана модель: $\frac{100}{100-Y_i} = \beta_1 + \beta_2 \left(\frac{1}{X_i}\right) + u_i$

Данные из таблицы 6.7:

| $Y_i$ | 86 | 79 | 76 | 69 | 65 | 62 | 52 | 51 | 51 | 48 |
|-------|----|----|----|----|----|----|----|----|----|----|
| $X_i$ | 3  | 7  | 12 | 17 | 25 | 35 | 45 | 55 | 70 | 120|

### Шаг 1: Преобразуем переменные

Рассчитаем значения для преобразованных переменных:
- $Y_i^* = \frac{100}{100-Y_i}$
- $X_i^* = \frac{1}{X_i}$

| $i$ | $Y_i$ | $X_i$ | $Y_i^* = \frac{100}{100-Y_i}$ | $X_i^* = \frac{1}{X_i}$ |
|-----|-------|-------|------------------------------|------------------------|
| 1   | 86    | 3     | 7.1429                      | 0.3333                |
| 2   | 79    | 7     | 4.7619                      | 0.1429                |
| 3   | 76    | 12    | 4.1667                      | 0.0833                |
| 4   | 69    | 17    | 3.2258                      | 0.0588                |
| 5   | 65    | 25    | 2.8571                      | 0.0400                |
| 6   | 62    | 35    | 2.6316                      | 0.0286                |
| 7   | 52    | 45    | 2.0833                      | 0.0222                |
| 8   | 51    | 55    | 2.0408                      | 0.0182                |
| 9   | 51    | 70    | 2.0408                      | 0.0143                |
| 10  | 48    | 120   | 1.9231                      | 0.0083                |

### Шаг 2: Оценка параметров регрессии

Теперь мы можем оценить модель $Y_i^* = \beta_1 + \beta_2 X_i^* + u_i$.

Для оценки параметров используем формулы МНК:

$\beta_2 = \frac{\sum (X_i^* - \bar{X^*})(Y_i^* - \bar{Y^*})}{\sum (X_i^* - \bar{X^*})^2}$

$\beta_1 = \bar{Y^*} - \beta_2 \bar{X^*}$

Где:
- $\bar{X^*}$ - среднее значение $X_i^*$
- $\bar{Y^*}$ - среднее значение $Y_i^*$

Рассчитываем средние значения:
- $\bar{X^*} = 0.0750$
- $\bar{Y^*} = 3.2874$

Для расчета $\beta_2$ необходимо составить вспомогательную таблицу:

| $i$ | $X_i^*$ | $Y_i^*$ | $(X_i^* - \bar{X^*})$ | $(Y_i^* - \bar{Y^*})$ | $(X_i^* - \bar{X^*})(Y_i^* - \bar{Y^*})$ | $(X_i^* - \bar{X^*})^2$ |
|-----|---------|---------|----------------------|----------------------|----------------------------------------|------------------------|
| 1   | 0.3333  | 7.1429  | 0.2583               | 3.8555               | 0.9961                                 | 0.0667                 |
| 2   | 0.1429  | 4.7619  | 0.0679               | 1.4745               | 0.1001                                 | 0.0046                 |
| ... | ...     | ...     | ...                  | ...                  | ...                                    | ...                    |
| 10  | 0.0083  | 1.9231  | -0.0667              | -1.3643              | 0.0910                                 | 0.0044                 |

Из расчетов получаем:
- Сумма $(X_i^* - \bar{X^*})(Y_i^* - \bar{Y^*}) = 1.5821$
- Сумма $(X_i^* - \bar{X^*})^2 = 0.0880$

Следовательно:
- $\beta_2 = \frac{1.5821}{0.0880} = 17.9784$
- $\beta_1 = 3.2874 - 17.9784 \times 0.0750 = 1.9390$

Таким образом, оцененная модель имеет вид:
$\frac{100}{100-Y_i} = 1.9390 + 17.9784 \left(\frac{1}{X_i}\right)$

### Шаг 3: Статистические показатели модели

1. **Коэффициент детерминации (R²):**
   
   R² = 0.9768, что означает, что модель объясняет около 97.68% вариации зависимой переменной.

2. **Стандартные ошибки:**
   
   - Стандартная ошибка для $\beta_1$: 0.1126
   - Стандартная ошибка для $\beta_2$: 0.9875

3. **t-статистики:**
   
   - t-статистика для $\beta_1$: 17.22 (p-значение < 0.0001)
   - t-статистика для $\beta_2$: 18.21 (p-значение < 0.0001)

4. **F-статистика:**
   
   F = 331.56 (p-значение < 0.0001)

### Шаг 4: Интерпретация результатов

1. **Значимость модели:** Высокое значение F-статистики и низкое p-значение указывают на статистическую значимость модели в целом.

2. **Значимость коэффициентов:** Оба коэффициента статистически значимы (p < 0.05).

3. **Качество подгонки:** Высокий коэффициент детерминации (R² = 0.9768) свидетельствует о хорошей объяснительной способности модели.

4. **Экономическая интерпретация:**
   - Когда $X_i$ стремится к бесконечности, $1/X_i$ стремится к нулю, и $Y_i^*$ приближается к $\beta_1 = 1.9390$.
   - Это означает, что $Y_i$ приближается к значению около 48.4%, когда $X_i$ становится очень большим.
   - Коэффициент $\beta_2 = 17.9784$ показывает, насколько изменяется преобразованная зависимая переменная при изменении преобразованной независимой переменной.

## Задание 2: Модель CES-производственной функции

Дана модель: $\log\left(\frac{V}{L}\right) = \log \beta_1 + \beta_2 \log W + u$

где:
- $(V/L)$ = добавленная стоимость на единицу труда
- $L$ = трудовой вклад
- $W$ = ставка реальной заработной платы
- $\beta_2$ = эластичность замещения между трудом и капиталом

### Шаг 1: Оценка параметров регрессии

Используем метод наименьших квадратов для оценки параметров модели $\log(V/L) = \log \beta_1 + \beta_2 \log W + u$.

Для оценки параметров используем формулы МНК:

$\beta_2 = \frac{\sum (\log W_i - \overline{\log W})(\log(V/L)_i - \overline{\log(V/L)})}{\sum (\log W_i - \overline{\log W})^2}$

$\log \beta_1 = \overline{\log(V/L)} - \beta_2 \times \overline{\log W}$

Рассчитываем средние значения:
- $\overline{\log W} = 3.0234$
- $\overline{\log(V/L)} = 3.5800$

Из расчетов получаем:
- $\beta_2 = 1.3338$
- $\log \beta_1 = 3.5800 - 1.3338 \times 3.0234 = -0.4515$
- $\beta_1 = e^{-0.4515} = 0.6368$

Таким образом, оцененная модель имеет вид:
$\log\left(\frac{V}{L}\right) = -0.4515 + 1.3338 \log W$
или
$\log\left(\frac{V}{L}\right) = \log 0.6368 + 1.3338 \log W$

### Шаг 2: Проверка статистической значимости эластичности замещения

Нам необходимо проверить гипотезу о том, что эластичность замещения ($\beta_2$) статистически не отличается от 1.

Нулевая гипотеза: $H_0: \beta_2 = 1$
Альтернативная гипотеза: $H_1: \beta_2 \neq 1$

Для проверки используем t-тест:

$t = \frac{\beta_2 - 1}{SE(\beta_2)}$

где $SE(\beta_2)$ - стандартная ошибка коэффициента $\beta_2$.

Рассчитываем стандартную ошибку:
- $SE(\beta_2) = 0.2519$

Таким образом:
- $t = \frac{1.3338 - 1}{0.2519} = 1.3252$

При уровне значимости $\alpha = 0.05$ и степенях свободы $df = n - 2 = 15 - 2 = 13$, критическое значение t-статистики равно $t_{критическое} = \pm 2.1604$.

Поскольку $|t| = 1.3252 < t_{критическое} = 2.1604$, мы не можем отвергнуть нулевую гипотезу.

### Шаг 3: Дополнительные статистические показатели

1. **Коэффициент детерминации:**
   
   $R^2 = 0.5764$, что означает, что модель объясняет около 57.64% вариации зависимой переменной.

2. **F-статистика:**
   
   $F = 17.67$ (p-значение = 0.0010)

3. **p-значение для $\beta_2 = 1$:**
   
   $p = 0.2084 > 0.05$, что подтверждает, что мы не можем отвергнуть гипотезу о том, что $\beta_2 = 1$.

### Шаг 4: Интерпретация результатов

1. **Значимость модели:** Модель в целом статистически значима (F = 17.67, p < 0.05).

2. **Эластичность замещения:** Оцененная эластичность замещения между трудом и капиталом составляет 1.3338.

3. **Проверка гипотезы:** Мы не можем отвергнуть гипотезу о том, что истинное значение эластичности равно 1 (p = 0.2084 > 0.05).

4. **Экономическая интерпретация:**
   - Коэффициент $\beta_2 = 1.3338$ показывает эластичность замещения между трудом и капиталом.
   - Значение больше 1 указывает на высокую степень замещаемости факторов производства.
   - Однако, поскольку статистически мы не можем отвергнуть гипотезу о том, что $\beta_2 = 1$, это свидетельствует о том, что эластичность замещения может соответствовать функции Кобба-Дугласа.
   - CES-производственная функция при $\beta_2 = 1$ переходит в функцию Кобба-Дугласа.

## Заключение

1. В первой модели мы имеем высокий коэффициент детерминации (R² = 0.9768), что указывает на хорошую объяснительную способность модели. Оба коэффициента статистически значимы.

2. Во второй модели оцененная эластичность замещения между трудом и капиталом составляет 1.3338, но статистически мы не можем отвергнуть гипотезу о том, что эластичность равна 1. Это согласуется с функцией Кобба-Дугласа, что является важным результатом для понимания производственной технологии в рассматриваемых отраслях.
