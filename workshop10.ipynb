
## Задача 10.2


### a) Можем ли мы оценить три неизвестных?

**Ответ:**

Нет, нельзя.

**Объяснение:**

- Между \( X_2 \) и \( X_3 \) существует **строгая линейная зависимость**:
  
  \[
  X_3 = 2X_2 - 1
  \]
  
- Следовательно, существует **идеальная мультиколлинеарность**.
- В этом случае матрица \( X'X \) становится вырожденной (её определитель равен нулю), что делает оценивание с помощью МНК невозможным.

### b) Какие линейные комбинации параметров можно оценить?

**Ответ:**

Можно оценить **линейные комбинации** \(\beta_2\) и \(\beta_3\).

**Пример:**

Из зависимости \( X_3 = 2X_2 - 1 \), можно подставить \( X_3 \) через \( X_2 \) в модель:

\[
Y_i = \beta_1 + \beta_2 X_{2i} + \beta_3 (2X_{2i} - 1) + u_i
\]
\[
= (\beta_1 - \beta_3) + (\beta_2 + 2\beta_3)X_{2i} + u_i
\]

Таким образом, можно оценить:

- Константу: \( \beta_1 - \beta_3 \)
- Коэффициент при \( X_2 \): \( \beta_2 + 2\beta_3 \)

Индивидуально \( \beta_2 \) и \( \beta_3 \) оценить нельзя.

## Задача 10.4

**Дано:** линейная зависимость:

\[
\lambda_1 X_{1i} + \lambda_2 X_{2i} + \lambda_3 X_{3i} = 0
\]

---

### Решение:

#### 1. Проверка мультиколлинеарности

По таблице видно:

- \( X_2 \) и \( X_3 \) связаны формулой \( X_3 = 2X_2 - 1 \).
Это значит, что \( X_2 \) и \( X_3 \) практически линейно зависимы.
Таким образом, мультиколлинеарность близка к идеальной.

#### 2. Корреляции

- Поскольку \( X_2 \) и \( X_3 \) идеально линейно зависимы, то коэффициент корреляции между ними равен почти \( \pm1 \).

#### 3. Коэффициенты детерминации

- \( R^2_{1,23} \), \( R^2_{2,13} \), \( R^2_{3,12} \) будут близки к 1.
- Например:
  \[
  R^2_{3,12} \approx 1
  \]
  (Поскольку \( X_3 \) выражается через \( X_2 \)).

---

### Заключение:

- В данной ситуации присутствует **идеальная мультиколлинеарность**.
- Следовательно, стандартная регрессия невозможна без удаления одной из линейно зависимых переменных.
