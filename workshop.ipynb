```python
df = pd.DataFrame(data)

countries = ["Canada", "France", "Germany", "Italy", "Japan", "UK", "US"]
for country in countries:
    # shift(1) references previous year's CPI
    df[f"{country}_inflation"] = (
        (df[country] - df[country].shift(1)) / df[country].shift(1)
    ) * 100

```

```python
plt.figure(figsize=(10, 6))
for country in countries:
    plt.plot(df["Year"], df[f"{country}_inflation"], label=country)

plt.title("Year-over-Year Inflation (Seven Industrialized Countries, 1973–1997)")
plt.xlabel("Year")
plt.ylabel("Inflation Rate (%)")
plt.grid(True)
plt.legend()
plt.show()
```

# Exercise 1  
1970s High Inflation
All seven countries experienced relatively high inflation in the 1970s—often reaching double‐digit levels. This was driven in part by:

Oil price shocks (1973, 1979) that sharply raised energy costs and overall prices.
Looser monetary policy in some nations, and widespread wage–price “spirals.”
Decline in the 1980s
By the early‐to‐mid 1980s, inflation was brought under control in most countries, generally trending down into the single‐digit range. Key contributing factors included:

Tighter monetary policy (for example, by the Federal Reserve in the U.S. and other central banks), meant to break inflationary expectations.
Slower growth and/or recessions that reduced aggregate demand.
Stability in the 1990s
In the late 1980s and through the 1990s, the major industrial economies settled into moderate, low single‐digit inflation. Central‐bank independence and inflation‐targeting frameworks (e.g., in Canada, the U.K., and later the European Central Bank model) helped keep inflation expectations in check.

Overall, the broad pattern is one of high and volatile inflation in the 1970s, followed by declining and eventually relatively stable inflation in the 1990s.

## 1.2(b)

From either a time‐series plot or a scatter plot, you typically see broad comovement—especially in the 1970s (when inflation peaked around the same time for all) and then again in the 1980s–1990s (when inflation trended downward almost everywhere). In many years, the U.S. rate is roughly “in the middle” of the seven, with countries like Italy and the U.K. often having more extreme swings.

Key observations:

Joint Response to Global Shocks
All these countries faced the same global oil‐price shocks (e.g., in 1973–74 and 1979) and hence saw similar inflation surges.

Shared Downward Trend
By the mid‐1980s, most central banks had begun more restrictive monetary policies, and by the 1990s inflation rates converged to lower levels—again paralleling the U.S. experience.

Some Differences
Certain countries (Italy, and sometimes the U.K.) experienced larger swings than did the U.S., reflecting more volatile exchange rates, different policy regimes, or faster wage–price pass‐through.
## 1.2(C)
Even if you see that the six countries’ inflation rates tend to move up and down in the same direction (or are positively correlated) with the U.S., it does not necessarily mean that U.S. inflation causes theirs. Several points illustrate why:

Global Common Shocks
Large oil‐price changes, global demand shifts, and worldwide commodity cycles can push up (or push down) everyone’s price level at roughly the same time.

Similar Policy Frameworks
Many industrial economies follow like‐minded monetary and fiscal policies, responding in tandem to inflationary or recessionary pressures. Thus, they move together not because the U.S. “forces” them to inflate, but because they face analogous conditions and policy choices.

Hence, similar inflation paths usually indicate they are subject to many of the same underlying influences, rather than a strict one‐way causal link from the U.S. to the rest.
