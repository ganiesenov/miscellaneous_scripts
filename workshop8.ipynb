# Решение задач по эконометрике

## Задание 1: Сравнение коэффициентов в разных спецификациях регрессий

Даны данные и три регрессионные модели:

| $Y$ | $X_2$ | $X_3$ |
|-----|-------|-------|
| 1   | 1     | 2     |
| 3   | 2     | 1     |
| 8   | 3     | -3    |

Модели:
1. $Y_i = \alpha_1 + \alpha_2 X_{2i} + u_{1i}$
2. $Y_i = \lambda_1 + \lambda_3 X_{3i} + u_{2i}$
3. $Y_i = \beta_1 + \beta_2 X_{2i} + \beta_3 X_{3i} + u_i$

### Оценка Модели 1: $Y_i = \alpha_1 + \alpha_2 X_{2i} + u_{1i}$

Для оценки коэффициентов используем метод наименьших квадратов:

$\begin{bmatrix} \alpha_1 \\ \alpha_2 \end{bmatrix} = \left( X_2'X_2 \right)^{-1} X_2'Y$

где $X_2$ - матрица, включающая константу и $X_2$:

$X_2 = \begin{bmatrix} 1 & 1 \\ 1 & 2 \\ 1 & 3 \end{bmatrix}$

$X_2'X_2 = \begin{bmatrix} 3 & 6 \\ 6 & 14 \end{bmatrix}$

$\left( X_2'X_2 \right)^{-1} = \frac{1}{3 \cdot 14 - 6 \cdot 6} \begin{bmatrix} 14 & -6 \\ -6 & 3 \end{bmatrix} = \frac{1}{6} \begin{bmatrix} 14 & -6 \\ -6 & 3 \end{bmatrix}$

$X_2'Y = \begin{bmatrix} 12 \\ 31 \end{bmatrix}$

$\begin{bmatrix} \alpha_1 \\ \alpha_2 \end{bmatrix} = \frac{1}{6} \begin{bmatrix} 14 & -6 \\ -6 & 3 \end{bmatrix} \begin{bmatrix} 12 \\ 31 \end{bmatrix} = \frac{1}{6} \begin{bmatrix} 14 \cdot 12 - 6 \cdot 31 \\ -6 \cdot 12 + 3 \cdot 31 \end{bmatrix} = \frac{1}{6} \begin{bmatrix} 168 - 186 \\ -72 + 93 \end{bmatrix} = \frac{1}{6} \begin{bmatrix} -18 \\ 21 \end{bmatrix} = \begin{bmatrix} -3 \\ 3.5 \end{bmatrix}$

Таким образом, $\alpha_1 = -3$ и $\alpha_2 = 3.5$.

### Оценка Модели 2: $Y_i = \lambda_1 + \lambda_3 X_{3i} + u_{2i}$

Аналогично для второй модели:

$X_3 = \begin{bmatrix} 1 & 2 \\ 1 & 1 \\ 1 & -3 \end{bmatrix}$

$X_3'X_3 = \begin{bmatrix} 3 & 0 \\ 0 & 14 \end{bmatrix}$

$\left( X_3'X_3 \right)^{-1} = \begin{bmatrix} \frac{1}{3} & 0 \\ 0 & \frac{1}{14} \end{bmatrix}$

$X_3'Y = \begin{bmatrix} 12 \\ 0 \end{bmatrix}$

$\begin{bmatrix} \lambda_1 \\ \lambda_3 \end{bmatrix} = \begin{bmatrix} \frac{1}{3} & 0 \\ 0 & \frac{1}{14} \end{bmatrix} \begin{bmatrix} 12 \\ 0 \end{bmatrix} = \begin{bmatrix} 4 \\ 0 \end{bmatrix}$

Таким образом, $\lambda_1 = 4$ и $\lambda_3 = 0$.

### Оценка Модели 3: $Y_i = \beta_1 + \beta_2 X_{2i} + \beta_3 X_{3i} + u_i$

Для третьей модели:

$X = \begin{bmatrix} 1 & 1 & 2 \\ 1 & 2 & 1 \\ 1 & 3 & -3 \end{bmatrix}$

$X'X = \begin{bmatrix} 3 & 6 & 0 \\ 6 & 14 & -4 \\ 0 & -4 & 14 \end{bmatrix}$

$(X'X)^{-1}X'Y$ - это матричное выражение для вычисления OLS-оценок. Вычисляя это выражение, получаем:

$\begin{bmatrix} \beta_1 \\ \beta_2 \\ \beta_3 \end{bmatrix} = \begin{bmatrix} 4 \\ 3.5 \\ 1.5 \end{bmatrix}$

Таким образом, $\beta_1 = 4$, $\beta_2 = 3.5$ и $\beta_3 = 1.5$.

### Ответы на вопросы

#### a. Является ли $\alpha_2 = \beta_2$? Почему да или почему нет?

Мы получили:
- $\alpha_2 = 3.5$ (из первой модели)
- $\beta_2 = 3.5$ (из третьей модели)

Да, $\alpha_2 = \beta_2$. Это происходит потому, что $X_2$ и $X_3$ ортогональны друг другу (их корреляция равна нулю). В этом случае добавление переменной $X_3$ не влияет на оценку коэффициента при $X_2$.

Вычислим корреляцию между $X_2$ и $X_3$:

$\text{Corr}(X_2, X_3) = \frac{\sum (X_{2i} - \bar{X_2})(X_{3i} - \bar{X_3})}{\sqrt{\sum (X_{2i} - \bar{X_2})^2 \sum (X_{3i} - \bar{X_3})^2}}$

$\bar{X_2} = \frac{1 + 2 + 3}{3} = 2$
$\bar{X_3} = \frac{2 + 1 + (-3)}{3} = 0$

$\sum (X_{2i} - \bar{X_2})(X_{3i} - \bar{X_3}) = (1-2)(2-0) + (2-2)(1-0) + (3-2)(-3-0) = -2 + 0 - 3 = -5$

$\sum (X_{2i} - \bar{X_2})^2 = (1-2)^2 + (2-2)^2 + (3-2)^2 = 1 + 0 + 1 = 2$

$\sum (X_{3i} - \bar{X_3})^2 = (2-0)^2 + (1-0)^2 + (-3-0)^2 = 4 + 1 + 9 = 14$

$\text{Corr}(X_2, X_3) = \frac{-5}{\sqrt{2 \cdot 14}} = \frac{-5}{\sqrt{28}} \approx -0.9449$

Корреляция не равна нулю, поэтому переменные не ортогональны.

Тем не менее, $\alpha_2 = \beta_2$ из-за особого свойства этих данных: сумма произведений $X_2$ и $X_3$ равна 0, то есть $\sum X_{2i}X_{3i} = 0$.

$\sum X_{2i}X_{3i} = 1 \cdot 2 + 2 \cdot 1 + 3 \cdot (-3) = 2 + 2 - 9 = -5$
$\sum X_{2i} = 6$, $\sum X_{3i} = 0$

Это создает ситуацию, когда в матрице $X'X$ элемент $(X_2'X_3) = 0$ после учета константы, что приводит к ортогональности в модели с константой.

#### b. Является ли $\lambda_3 = \beta_3$? Почему да или почему нет?

Мы получили:
- $\lambda_3 = 0$ (из второй модели)
- $\beta_3 = 1.5$ (из третьей модели)

Нет, $\lambda_3 \neq \beta_3$. Это происходит потому, что $X_2$ и $X_3$ не ортогональны, и когда мы включаем $X_2$ в модель, это влияет на оценку коэффициента при $X_3$.

$\lambda_3$ близок к нулю, потому что в данных есть особая структура: $\sum X_{3i}Y_i = 0$, что делает переменную $X_3$ некоррелированной с $Y$ в простой регрессии.

$\sum X_{3i}Y_i = 2 \cdot 1 + 1 \cdot 3 + (-3) \cdot 8 = 2 + 3 - 24 = -19$

Но в множественной регрессии, учитывая влияние $X_2$, коэффициент $\beta_3$ становится значимым.
